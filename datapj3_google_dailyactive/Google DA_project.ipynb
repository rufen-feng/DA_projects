# EDA

### we know enough information of this data set now, we can begin to do EDA

```python
active_pivot=active.groupby(active['Id'])['TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes'].mean()
active_pivot
```

```python
(active_pivot['LoggedActivitiesDistance']== 0).value_counts()/active_pivot.shape[0]
```

### the total distance and the tracker distance is the same, so we can delete one of them, and we can see the procuct can track the exact number of active distance. but as we look at the logged one, we lost nearly 88% of the data, so we need to back to our producut to see if there is problem with the hardware, or anything wrong with the customers' behavior

```python
active_pivot = active_pivot.drop(columns=['TrackerDistance', 'LoggedActivitiesDistance'])
active_pivot.head()
```

```python
active_pivot['highdis%'] = (active_pivot['VeryActiveDistance']/active_pivot['TotalDistance']).round(2)
active_pivot['moderdis%'] = (active_pivot['ModeratelyActiveDistance']/active_pivot['TotalDistance']).round(2)
active_pivot['lightdis%'] = (active_pivot['LightActiveDistance']/active_pivot['TotalDistance']).round(2)
active_pivot['sedendis%'] = (active_pivot['SedentaryActiveDistance']/active_pivot['TotalDistance']).round(2)
```

```python
active_pivot.head(50)
```

```python
active_pivot.shape
```

```python
active_pivot.reset_index(drop=False, inplace=True)
```

```python
active2 = active_pivot[['Id','highdis%','moderdis%','lightdis%','sedendis%']]
active2
```

```python
ax = active2.plot(kind='bar', x='Id', y=['highdis%', 'moderdis%', 'lightdis%', 'sedendis%'], figsize=(14, 7), color=("#7FFFD4","#FFB6C1","#808080","#FFFF00"))

# Customize the plot properties
plt.title('Status percent')
plt.xlabel('Id')
plt.ylabel('Percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.show()
```

### the sedentary catagory disappears in this plot, so the plot is not reliable for some product concerning sedentary time. However, we still find fo rmost of the users, lighttime is the largest,  second the moderdis, finally the highdis. This is a common phenomenon of group consumption.

```python
df = active2[['highdis%', 'moderdis%', 'lightdis%', 'sedendis%']].mean()
df.plot(kind='bar', color = "#808080", figsize=(10,6))
```

### some people are unique, they spend most of the time on highdis, for these users, we can track their behavior, and recommend unique products, as they are actively excersize.

```python
active_users = active2[active2['highdis%']>0.4]
active_users
```

```python
selected=['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']
total_activity_minutes = active_pivot[selected].sum(axis=1)
active_pivot['activemin%'] = (active_pivot['VeryActiveMinutes']/total_activity_minutes).round(2)
active_pivot['fairmin%'] = (active_pivot['FairlyActiveMinutes']/total_activity_minutes).round(2)
active_pivot['lightmin%'] = (active_pivot['LightlyActiveMinutes']/total_activity_minutes).round(2)
active_pivot['sedentarymin%'] = (active_pivot['SedentaryMinutes']/total_activity_minutes).round(2)
```

```python
active_pivot.head(2)
```

```python
active_pivot.columns
```

```python
active3=active_pivot[['Id','activemin%', 'fairmin%', 'lightmin%', 'sedentarymin%']]
active3.head(5)
```

```python
ax = active3.plot(kind='bar', x='Id', y=['activemin%', 'fairmin%', 'lightmin%', 'sedentarymin%'], figsize=(14, 7), 
                  color=("#7FFFD4","#FFFF00","#808080","#FFB6C1"))

# Customize the plot properties
plt.title('Status percent')
plt.xlabel('Id')
plt.ylabel('Percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.show()
```

```python
df2=active3[['activemin%', 'fairmin%', 'lightmin%', 'sedentarymin%']].mean()
df2.plot(kind='bar',color="#808080",figsize=(10,6))
```
